2. Find The Number Of Unique Jobs Titles


```python
num_unique_jobs = df['JobTitle'].nunique()
print(f"Number of unique job titles: {num_unique_jobs}")
```

Documentation

This output means there are 1109 different job titles in the dataset. In other words, there are 1109 unique entries within the 'JobTitle' column of the DataFrame. This information provides insight into the diversity of job roles represented in the dataset.

3. Find The Total Number Of Job Titles Contain captain


```python
captain_count = df['JobTitle'].str.contains('captain', case=False).sum()
print(f"Number of job titles containing 'captain': {captain_count}")
```

Documentation

This output means that there are 410 job titles in the dataset that contain the word "captain" (in any case). This provides information about the prevalence of job roles related to "captain" within the dataset.

5. Find The Minimum,Maximum and Average BasePay


```python
minimum_basepay = df['BasePay'].min()
maximum_basepay = df['BasePay'].max()
average_basepay = df['BasePay'].mean()

print(f"Minimum BasePay: {minimum_basepay}")
print(f"Maximum BasePay: {maximum_basepay}")
print(f"Average BasePay: {average_basepay}")
```

Documentation

Minimum BasePay: -166.01: This indicates the lowest value present in the 'BasePay' column of the DataFrame. In this case, the minimum base pay is -166.01. A negative base pay could possibly represent deductions or specific situations within the dataset's context.
Maximum BasePay: 319275.01: This represents the highest value in the 'BasePay' column. Here, the maximum base pay is 319275.01.
Average BasePay: 67207.55842464653: This shows the average or mean value of all the base pay values in the DataFrame. The average base pay is approximately 67207.56.

10. Display Name Of The Person Having The Highest Basepay

11. Find The average Basepay of all Employees per Year


```python
average_basepay_per_year = df.groupby('Year')['BasePay'].mean()
print(average_basepay_per_year)
```

12. Find Average Basepay Of All Employee Per Job Title


```python
average_basepay_per_job = df.groupby('JobTitle')['BasePay'].mean()
print(average_basepay_per_job)
```
